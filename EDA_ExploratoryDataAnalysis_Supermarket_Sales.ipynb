```python
data["Rating"].skew()
```

Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable. 
Here are the three main characteristics of skewness in data distributions:

1. Positive Skew (Right Skew)
Shape: The right tail (higher values) of the distribution is longer or fatter than the left tail (lower values).
Skewness value: Greater than 0 (positive).

2. Negative Skew (Left Skew)
Shape: The left tail (lower values) is longer or fatter than the right tail (higher values).
Skewness value: Less than 0 (negative).

#### 3. Zero Skew (Symmetric Distribution) This is out case  the Data is Balenced
Shape: The distribution is symmetric, meaning both tails (left and right) are approximately the same length and shape.
Skewness value: Close to 0. The data is balanced

### 9. Display Daily Sales By Day Of The Week

```python
date_map = {
    0:"Monday",
    1:"Tuesday",
    2:"Wednesday",
    3:"Thursday",
    4:"Friday",
    5:"Saturday",
    6:"Sunday"
}
```

```python
data["day_of_week"] = data["Date"].dt.dayofweek.map(date_map)
```

```python
data.head(1)
```

```python
data["day_of_week"].value_counts().plot(kind="bar")
```

### 10. What Will Be The Highest Months For Sales

```python
months = {
    1:"January",
    2:"February",
    3:"March",
    4:"April",
    5:"May",
    6:"June", 
    7:"July",
    8:"August",
    9:"September",
    10:"October",
    11:"November",
    12:"December",
}
```

```python
data["Months"] = data["Date"].dt.month.map(months)
```

```python
data.head(1)
```

```python
data["Months"].value_counts().plot(kind="bar")
plt.show()
```
